```python
import pyspark.sql.functions as f
from pyspark.sql.types import StringType
from pyspark.sql.functions import trim, col
```

#read

```python
df=spark.table("databricks_bootcamp.bronze.prd_info")
```

```python
df.display()
```

#norlmilation

```python
df = df.withColumn(
    "prd_line",
    f.when(f.upper(f.trim(f.col("prd_line"))) == "R", "regular")
     .when(f.upper(f.trim(f.col("prd_line"))) == "S", "small")
     .when(f.upper(f.trim(f.col("prd_line"))) == "M", "medium")
     .when(f.upper(f.trim(f.col("prd_line"))) == "T", "trial")
     .otherwise("N/A")
)
```

```python
df=df.withColumn("prd_cost",f.when(f.col("prd_cost").isNull(),0).otherwise(f.col("prd_cost")))
```

```python
df.display()
```

#Rename columns 

```python
rename={
    "prd_id":"product_id",
    "prd_key":"product_key",
    "prd_nm":"product_name",
    "prd_cost":"product_cost",
    "prd_line":"product_line",
    "prd_start_dt":"product_start_date",
    "prd_end_dt":"product_end_date"
}
for old_name,new_name in rename.items():
    df=df.withColumnRenamed(old_name,new_name)
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("databricks_bootcamp.silver.proudect_info")
```
