```python
df=spark.table("databricks_bootcamp.bronze.sales_details")
```

```python
df.display()
```

```python
rename={
    "sls_ord_num":"order_id",
    "sls_prd_key":"product_key",
    "sls_cust_id":"customer_id",
    "sls_order_dt":"order_date",
    "sls_ship_dt":"ship_date",
    "sls_due_dt":"due_date",
    "sls_sales":"sales",
    "sls_quantity":"quantity",
    "sls_price":"price"
}
for old_name,new_name in rename.items():
    df=df.withColumnRenamed(old_name,new_name)
df.display()
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("databricks_bootcamp.silver.sales")
```
